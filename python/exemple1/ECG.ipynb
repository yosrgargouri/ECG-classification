```python
import matplotlib.pyplot as plt

import pandas as pd
data=pd.read_csv('C:/Users/USER/Google Drive/stage/python/ECG/samples.csv')
data
```

```python
fig,axes=plt.subplots(12, 1, figsize=(12,20))

leads=['I','II','III','avR','avL','avF','V1','V2','V3','V4','V5','V6']

for i, ax in enumerate(axes.flatten()):
    y=data.iloc[0:2000 ,i].values
    y=y/1000
    plt.ylim(-1.5,3)
    ax.plot(y,color=plt.cm.Paired(i/10.), label=leads[i])
    axes[i].set_ylabel('Amplitude')
    #axes[i].set_label(leads[i])
    plt.xlabel('Sapmle Point')
    axes[i].legend()
plt.savefig('C:/Users/USER/Google Drive/stage/python/ECG/ECG.pdf')
plt.show()
```
